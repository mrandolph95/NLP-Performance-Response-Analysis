```python
import pandas as pd
import langid
from langdetect import detect, DetectorFactory, LangDetectException
from lingua import Language, LanguageDetectorBuilder
```

```python
df = pd.read_csv('comments.csv')
```

```python
df.head()
```

```python
# Apply langid for language detection
def detect_langid(text):
    return langid.classify(str(text))[0]

df['langid'] = df['Comments'].apply(detect_langid)
```

```python
# LangDetect setup
DetectorFactory.seed = 0

def detect_langdetect(text):
    try:
        if isinstance(text, str):
            return detect(text)
        else:
            return "unknown"
    except LangDetectException:
        return "unknown"

# Apply langdetect
df['lang_detect'] = df['Comments'].apply(detect_langdetect)
```

```python
df.head()
```

```python
# Initialize Lingua language detector
detector = LanguageDetectorBuilder.from_all_languages().build()

# Function to detect language using Lingua
def detect_language_lingua(text):
    try:
        language = detector.detect_language_of(text)
        if language is None:
            return "unknown"
        else:
            return language.name
    except Exception as e:
        return "error"

# Apply Lingua language detection
df['Lingua'] = df['Comments'].apply(detect_language_lingua)
```
