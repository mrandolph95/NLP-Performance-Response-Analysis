```python
from langdetect import detect, DetectorFactory
from langdetect.lang_detect_exception import LangDetectException
```

```python
import langid
import pandas as pd
```

```python
k_df = pd.read_csv('/content/drive/MyDrive/kendrick_comment_data.csv')
```

```python
b_df = pd.read_csv('/content/drive/MyDrive/beyonce_comment_data.csv')
```

```python
k_df.head()
# Comments are incorrectly classified as non-english words (lines 0 and 3)
```

```python
b_df['lang'] = b_df['comment'].apply(lambda x: langid.classify(str(x))[0])
k_df['lang'] = k_df['comment'].apply(lambda x: langid.classify(str(x))[0])
```

```python
DetectorFactory.seed = 0

def detect_lang(text):
  try:
      if isinstance(text, str):
        return detect(text)
      else:
        return "unknown"
  except LangDetectException:
      return "unknown"
```

```python
b_df['lang_detect'] = b_df['comment'].apply(detect_lang)
k_df['lang_detect'] = k_df['comment'].apply(detect_lang)
```

```python
k_df.head()
```

```python
k_df = k_df.rename(columns={'lang':'langid'})
b_df = b_df.rename(columns={'lang':'langid'})
```

```python
k_df.to_csv('/content/drive/MyDrive/kendrick_comment_data_language_filter.csv', index=False)
b_df.to_csv('/content/drive/MyDrive/beyonce_comment_data_language_filter.csv', index=False)
```
